# LINE CHART BY FARAH HAMIDA KHAN:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y1 = np.sin(x)

z1=np.cos(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, linestyle="-")

plt.plot(x, z1, linestyle="--")

plt.title("Line Chart By FARAH")

plt.xlabel("Time")

plt.ylabel("Sales")

plt.show()
```

# BAR CHART BY FARAH:

```python
import matplotlib.pyplot as plt   

cat = ["Product A", "Product B", "Product C", "Product D"]
values = [27, 45, 69, 92]

plt.figure(figsize=(5, 5))

plt.bar(cat, values ,color="pink" ,edgecolor="skyblue" ,width=0.6)

plt.xticks(rotation=76,ha="left")
plt.title(" BAR Chart By FARAH ",fontsize=20)
plt.xlabel("Category",fontsize=20)
plt.ylabel("Sales")
plt.show()
```
