# Task 5: Building a Product Recommendation System

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
df = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
df.head(2)
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['overview'])

print(tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
# Create a reverse mapping of indices and movie titles
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

```

```python
def recommend(title, cosine_sim=cosine_sim, df=df):
    idx = indices.get(title)
    if idx is None:
        return f"Movie '{title}' not found in the dataset."

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]

    return df['title'].iloc[movie_indices].tolist()
```

```python
# Example 1
print("üìΩÔ∏è Recommendations for 'Avatar':")
print(recommend('Avatar'))

# Example 2
print("\nüìΩÔ∏è Recommendations for 'The Dark Knight':")
print(recommend('The Dark Knight'))

# Example 3
print("\nüìΩÔ∏è Recommendations for 'Interstellar':")
print(recommend('Interstellar'))
```
